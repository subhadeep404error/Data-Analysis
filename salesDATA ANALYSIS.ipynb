# EXPLATORY DATA ANALYSIS(EDA)

## CRITERIA--GENDER

```python

d.columns
```

```python
d1=sns.countplot(x='Gender',data=d,palette="Paired")

for bars in d1.containers:d1.bar_label(bars)

  #here we can see from the count plot that there are 7832 female has done the the maximum shopping than male 
```

```python
gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=gen,palette="rocket")
```

### we can see from the above graph that the amount spend by female in shopping is more than male.we can say that female does maximum shopping than male.

# 2nd Criteria--AGE




```python
d1=sns.countplot(x='Age Group',data=d,hue='Gender',palette='viridis')
for bars in d1.containers:d1.bar_label(bars)
plt.show()
```

### here we have ploted a count plot on the basis of age group because we want to see that which age group does maximum shopping and we have given hue parameter on gender.

```python
age=d.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x="Age Group",y="Amount",data = age)
plt.show()
```

#### when we ploted against age group vs amount it is seen that the maximum amount is spend by 26-35 age group. 

# 3rd criteria -- STATE


```python
d.columns
d.head(10)
```

```python
state=d.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)
plt.figure(figsize=(15,8))

sns.barplot(x='State',y='Orders',data=state,palette='rocket')
plt.show()
```

### i have plotted a graph against state vs orders and observed that uttar pradesh has placed maximum orders among other states.   

```python
amount_state=d.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
plt.figure(figsize=(15,8))
sns.barplot(x='State',y='Amount',data=amount_state)
plt.show()
```

### In this graph i have ploted amount verses state here also same result but we can see that in the previous graph hariyana has ordered less than kerela but when we ploted against amount we can see that haryana has ordered more than kerela,it is not in the top 10 list. 

# 3rd Criteria--Marital Status

```python
d.columns
```

```python
marital=sns.countplot(x="Marital_Status",data=d,palette="icefire")
for bars in marital.containers:marital.bar_label(bars)
plt.show()
 
```

# 4th Criteria--Occupation  

```python
d.columns
```

```python
d.head(10)
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x="Occupation",data=d)
plt.show()
```

# 6th criteria-- customer_name
